# Purpose: Convert the csv result from snowflake to a json file
## Functionality
- Create a df by reading the CSV file
- Convert the df to a list
  - Crteate a string by remove the first and the last character from the list 
- Using ast.literal_eval() to convert a string to a tuple    

```python
import json
# read csv
import pandas as pd
import ast 

```

```python
df = pd.read_csv("C:\\Users\\syue003\\Downloads\\json_bod.csv", skiprows = 1)

```

```python
df
```

```python
dl = list(df)
```

```python
dl
```

```python
dly= dl[0].split(":")[0]  # str: '{"table"'  len = 8
```

```python
#  Crteate a string by remove the first and the last character from the list 
#dl_x= dl[0][1:-1]   # type: str 
dl_x = dl[0][len(dly) + 2:-2]
```

```python
dl_x
```

```python
# Convert a string to a tuple
dl_y = ast.literal_eval(dl_x) # type: tuple
```

```python
dl_y
```

```python
# Convert as tuple to a list which will be used to create a json file
dl_z = list(dl_y)
```

```python
dl_z
```

## Create a json file

```python
j_dl = json.dumps(dl_z, indent=4)   # type = str
#j_dl = j_dl.replace("[", "{", 1)  # Replace the first [ with {
#j_dl = j_dl.replace("]", "}", -1) # Replace the last ] with } 
f = open('bod.json', 'w') # create a json file
print(j_dl, file=f)
f.close()
```
